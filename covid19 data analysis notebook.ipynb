### Let's Import the modules 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

### Importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
corona_dataset_csv = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head()
```

#### Let's check the shape of the dataframe

```python
corona_dataset_csv.shape
```

### Delete the useless columns

```python
df = corona_dataset_csv.drop(["Lat","Long"], axis = 1)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

### Aggregating the rows by the country

```python
df_aggregated = df.groupby("Country/Region").sum()
```

```python
df_aggregated.head()
```

```python
df_aggregated.shape
```

###  Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

```python
df_aggregated.loc["China"].plot()
df_aggregated.loc["Italy"].plot()
df_aggregated.loc["India"].plot()
plt.legend()
```

###  Calculating a good measure 
we need to find a good measure represented as a number, describing the spread of the virus in a country. 

```python
df_aggregated.loc['China'].plot()
```

```python
df_aggregated.loc['China'][:3].plot()
```

### Calculating the first derivative of the curve

```python
df_aggregated.loc['China'].diff().plot()     #this plot shows the change in infection rate day by day
```

### Finding maximum infection rate for China

```python
df_aggregated.loc["China"].diff().max()
```

```python
df_aggregated.loc["Italy"].diff().max()
```

```python
df_aggregated.loc["India"].diff().max()
```

### find maximum infection rate for all of the countries. 

```python
countries = list(df_aggregated.index)
```

```python
max_infection_rates = []
for c in countries:
    max_infection_rates.append(df_aggregated.loc[c].diff().max())
df_aggregated["max_infection_rates"] = max_infection_rates
```

```python
df_aggregated.head()
```

### Creating a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(df_aggregated["max_infection_rates"])
```

```python
corona_data.head()
```

### Plotting GDP vs maximum Infection rate

```python
x = data["GDP per capita"]
y = data["max_infection_rates"]
sns.scatterplot(x,y)          #we can see from  figure that this figure does not show enough details. So, we will do log scaling here
```

```python
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))

#This gives a very interesting result that people who are living in more developed countries
#are more prone to getting infection of corona virus as compared to people who are living in less developed countries
#But, you can also conclude that this is because of more testkits in more developed countries
```
